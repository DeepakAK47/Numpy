# Numpy array is faster than python list

```python
import time
```

```python
# WAP to create two list and apply addition operation over it and determine the time
x = range(100000)
y = range(100000,200000)

start_time = time.time()
c = [x+y for x,y in zip(x,y)]
end_time = time.time()

time_taken = end_time - start_time
time_taken
```

```python
# WAP to create two numpy array and apply addition operation over it and determine the time
x = np.arange(100000)
y = np.arange(100000,200000)

start_time = time.time()
c = x+y
end_time = time.time()

time_taken = end_time - start_time
time_taken
```
