```python
# findind the maximum of rows
arr  = np.array([[1,2,3,4,5],[9,7,6,54,3]])
max = np.max(arr,axis=1)
print(max)
```

```python
# findind the maximum of column
arr  = np.array([[1,2,3,4,5],[9,7,6,54,3]])
max = np.max(arr,axis=0)
print(max)
```
