```python
import matplotlib.pyplot as plt
import numpy as np

x = [10,20,30,40,50,60];
y1 = [i*2 for i in x]
y2 = [i**3 for i in x]
y3 = [np.sqrt(i) for i in x]

plt.subplot(2,2,1)
plt.plot(y1,x)
plt.title("First plot")

plt.subplot(2,2,2)
plt.plot(y2,x)
plt.title("Second plot")

plt.subplot(2,2,3)
plt.plot(y3,x)
plt.title("Third plot")

plt.subplot(2,2,4)
plt.plot(y3,x)
plt.title("Fourth plot")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [10,20,30,40,50,60]
y1 = [i*2 for i in x]
y2 = [i**3 for i in x]
y3 = [np.sqrt(i) for i in x]

fig,ax = plt.subplots(2,2)

ax[0][0].plot(x,y1,label=["x","y1"])
ax[0][0].set_title("Firts plot")

ax[0][1].plot(x,y2,label=["x","y2"])
ax[0][1].set_title("Second plot")

ax[1][0].plot(x,y3,label=["x","y3"])
ax[1][0].set_title("Third plot")

ax[1][1].plot(x,y3,label=["x","y3"])
ax[1][1].set_title("Fourth plot")
fig.tight_layout()
fig.show()
```

```python
# Temperature over cities
import matplotlib.pyplot as plt
import numpy as np
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]

fig,ax = plt.subplots(2,2)

count = 0
for i in range(2):
    for j in range(2):
        ax[i][j].plot(days,temperatures[count],marker='o')
        ax[i][j].grid(True)
        ax[i][j].set_title(cities[count])
        count = count+1

fig.suptitle("Temperature evaluation of different cities")
fig.supxlabel("Days")
fig.supylabel("Degree Celcious")
fig.tight_layout()
fig.show()
```
